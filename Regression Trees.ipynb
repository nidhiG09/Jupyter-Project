```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%203/data/real_estate_data.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
#  there are rows with missing values which we will deal with in pre-processing
data.isna().sum()
```

```python
# Data Pre-Processing
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
# Lets split the dataset into our features and what we are predicting (target)
X = data.drop(columns=["MEDV"])
Y = data["MEDV"]
```

```python
X.head()
```

```python
Y.head()
```

```python
# Finally lets split our data into a training and testing dataset using train_test_split from sklearn.model_selection
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.2, random_state=1)
```

```python
# Create Regression Tree
# DecisionTreeRegressor object, setting the criterion parameter to mse for Mean Squared Error
regression_tree = DecisionTreeRegressor(criterion = 'squared_error')
```

```python
# Training                                  # The error occurs because the criterion parameter in the DecisionTreeRegressor no longer supports 'mse'. Instead, you should use 'squared_error', which is equivalent to the old 'mse'.
regression_tree.fit(X_train, Y_train)
```

```python
# Evaluation
regression_tree.score(X_test, Y_test)
```

```python
prediction = regression_tree.predict(X_test)
print("$", (prediction - Y_test).abs().mean()*1000)
```

```python
# Exercise
regression_tree = DecisionTreeRegressor(criterion = 'absolute_error')
regression_tree.fit(X_train, Y_train)
print(regression_tree.score(X_test, Y_test))
prediction = regression_tree.predict(X_test)
print("$", (prediction - Y_test).abs().mean()*1000)
```
