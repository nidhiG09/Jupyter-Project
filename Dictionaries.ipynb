```python
# Create a sample dictionary
release_year_dict = {"Thriller": "1982", "Back in Black": "1980", \
                    "The Dark Side of the Moon": "1973", "The Bodyguard": "1992", \
                    "Bat Out of Hell": "1977", "Their Greatest Hits (1971-1975)": "1976", \
                    "Saturday Night Fever": "1977", "Rumours": "1977"}
release_year_dict
```

```python
# Get value by keys
release_year_dict["Thriller"]
```

```python
# Get all the keys in dictionary
release_year_dict.keys()
```

```python
# Get all the values in dictionary
release_year_dict.values()
```

```python
# Append value with key into dictionary
release_year_dict["Graduation"] = "2007"
release_year_dict
```

```python
# Delete entries by key
del(release_year_dict['Thriller'])
del(release_year_dict['Graduation'])
release_year_dict
```

```python
# Verify the key is in the dictionary
"The Bodyguard" in release_year_dict
```

## Scenario:Inventory Store

```python
# Task-1 Create an empty dictionary
inventory = {}
```

```python
# Task-2 Store the first product details in variable
Product_Name= "Mobile phone"
Product_Quantity= 5
Product_price= 20000
Product_Release_Year= 2020
```

```python
# Task-3 Add the details in inventory
inventory["Product_Name"] = Product_Name
inventory["Product_Quantity"] = Product_Quantity
inventory["Product_price"] = Product_price
inventory["Product_Release_Year"] = Product_Release_Year
```

```python
# Task-4 Store the second product details in a variable.¶
Product_Name1 = "Laptop"
Product_Quantity1 = 10
Product_price1 = 50000
Product_Release_Year1 = 2023
```

```python
# Task-5 Add the item detail into the inventory.
inventory["Product_Name1"] = Product_Name1
inventory["Product_Quantity1"] = Product_Quantity1
inventory["Product_price1"] = Product_price1
inventory["Product_Release_Year1"] = Product_Release_Year1
```

```python
# Task-6 Display the Products present in the inventory
print(inventory)
```

```python
# Task-7 Check if ProductNo1_releaseYear and ProductNo2_releaseYear is in the inventory¶
"Product_Release_Year" in inventory
```
