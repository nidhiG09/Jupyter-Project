```r
import requests
import pandas as pd

# Perform GET request
spacex_url = "https://api.spacexdata.com/v4/launches/past"
response = requests.get(spacex_url)

# Convert the JSON response to a dataframe
data = response.json()
df = pd.json_normalize(data)

# Extract rocket names from the 'rocket' column
rocket_names = []
for rocket_id in df['rocket']:
    if rocket_id:
        response = requests.get(f"https://api.spacexdata.com/v4/rockets/{rocket_id}")
        rocket_info = response.json()
        rocket_names.append(rocket_info['name'])
    else:
        rocket_names.append(None)

df['rocket_name'] = rocket_names

# Filter out Falcon 1 launches
df_falcon9 = df[df['rocket_name'] == 'Falcon 9']

# Count the number of Falcon 9 launches
num_falcon9_launches = len(df_falcon9)
print(num_falcon9_launches)
```
